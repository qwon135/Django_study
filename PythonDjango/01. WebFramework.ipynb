# 1. WebFramework

### 웹사이트의 작동방식

방문자가 웹 사이트의 서버에 요청을 보내면 서버는 필요한 정보를 방문자에게 보내준다.  
이때 사용자(클라이언트)가 하는 행동을 요청(request)라고 하며 서버가 이 요청에 응하는 동작을 응답(response)라고 한다.  
일반 적으로 사용자는 웹 브라우저에 URL을 입력해 서버로 요청을 전달하며, 서버는 미리 정해둔 규칙에 따라 URL을 해석하고 응답한다.   
요청에는 정보를 가져오는것 뿐만 아니라 서버에 내용을 저장하건 삭제하는것도 포함된다.(회원가입, 댓글 삭제 등)

### 웹 프레임 워크란?

웹 사이트는 request를 서버에서 처리하고, 결과를 볼 수 있는 형태로 만들어 response 하는 기능이 필요하다.  
불특정 다수에게 노출되는 만큼 성능과 보안 측면 또한 신경을 써야한다. 이런 복잡한 기능을 쉽게 만들 수 있게 도와주는 도구가 바로 웹 프레임워크이다.  
웹 프레임 워크는 웹 개발에 필요한 공통 요소들을 미리 개발해 놓은 상태로 제공한다.(회원가입, 탈퇴, 게시물 작성,수정,삭제 등)  
즉, 웹 프레임 워크를 사용하면 이미 개발 된 기능을 가져와 쉽게 추가할 수 있으며 많은 사람들이 사용한 만큼 성능과 보안 측면에서도 검증되어있다.

### 장고의 장점

#### 파이썬은 쉽다.
장고의 기반언어인 파이썬은 다른 언어에 비해 배우기 쉬우며 활용 범위가 넓다. 조금만 배우면 웹 크롤링을 이용한 데이터 수집, 업무 자동화에 활용이 가능하며 데이터 분석과 AI분야에도 파이썬 기반 라이브러리들을 적극 활용할 수 있다. 이러한 기능을 가진 파이썬으로 프로젝트를 할 때 웹 기반으로 발전 시킬 수 있다.
#### 웹 개발 진입 장벽을 낮출 수 있다.
처음 웹 개발을 할때 어려움은 처음 접하는 개념이 많다는 것이다. 컴퓨터가 서버의 기능을 할 수 있게 아파치나 Nginx 등의 웹서버 소프트웨어를 설치하거나, 방문자 정보 활용에 필요한 데이터베이스도 알아야한다. 장고는 파이썬 명령어로 서버를 실행할 수 있으며, 데이터베이스를 설치할 필요가 없다. 물론 전문적인 웹 서버 sw나 db를 사용하고 싶을 때도활용할 수 있다.
#### 관리 페이지와 보안기능을 기본제공한다.
장고는 프로젝트를 생성하는 순간 관리자 페이지를 자동으로 만들어주며, 웹사이트의 데이터베이스 구조를 쉽게 파악할 수 있다. 장고의 보안 가이드를 따라 웹개발을 하면 보안 실수를 할 가능성이 줄어든다.
