## (1) 웹 사이트 구조 구상하기
우선 '나'를 소개하는 웹 사이트인 블로그 겸 포트폴리오 사이트를 만들어 보자.

### 어떤 웹 사이트를 만들것인가?
만들 웹 페이지는 크게 3가지로 구성되어 있다.
#### Landing page(대문 페이지)
랜딩(대문) 페이지는 웹 사이트에 접속할 때 가장 처음 나타나는 대문 역할을 한다. 사이트의 정체성을 표현할 수 있는 문구나 이미지를 담고, 같이 일했던 사람의 평을 담으면 신뢰도를 높일 수도 있다.  
#### blog page
블로그 페이지는 말 그대로 평소에 공부한 내용이나 살면서 느낀점, 일기등을 기록하는 공간이다. 블로그 는 크게 두가지 요소로 구상한다.  
* Postlist page : 우선 블로그의 포스트를 목록 형태로 보여주는 목록 페이지를 만든다.
* Postdetail page : 방문자가 마음에 드는 포스트를 발견해 클릭한다면 포스트 전체 내용과 댓글을 보여주는 포스트 상세 페이지  

#### 자기소개 페이지
나에 대해 소개하고, 그동한 작업했던 내용을 간단한 이미지와 글로 보여주는 공간이다.

## (3) Pycharm 실습환경 설정
파이썬 전용 IDE중 하나인 Pycharm을 사용해보자.

### 파이참으로 가상환경 만들기
#### 프로젝트 폴더 지정하기
Pycharm을 실행 시킨 후 my_blog폴더를 연다.
#### 가상환경 만들기
이제 프로젝트를 위한 가상환경을 만든다. 장고를 비롯한 웹 사이트를 만드는데 필요한 외부라이브러리는 가상환경에 설치하는 것이 좋다. 가상환경을 만드는 방법은 터미널에 pyvenv, virtualenv 명령어를 입력하는 등의 방법이 있지만 파이썬에서 제공하는 기능을 이용하자.  

File>Settings에 들어가 my_blog폴더에 Python Interpreter를 클릭하고 Add를 누르자. 일련의 과정을 거쳐 my_blog 폴더에 venv 폴더가 설정되어있다. Ok를 눌러 새로운 가상환경을 venv에 생성해주자. 
> 가상환경을 사용하는 이유  
> 한 PC안에 웹 사이트 개발 이외에 다른 A 프로젝트를 별도로 진행하고 있는 중이라고 가정해보자. 만약 웹 사이트 개발 프로젝트를 위해 외부 모듈을 설치하였는데 파이썬 전체 개발 환경에 영향을 준다면 관계 없는 프로젝트 A에 의도치않은 영향을 끼질 수 있다. 가상환경은 이런 불상사를 예방해준다.

####  .gitignore에 가상환경 폴더(venv)추가하기
왼쪽 네비게이션 창을 보니 venv 폴더가 생성되었다. 처음에는 다른파일이나 폴더와 달리 흐릿하게 보이는데 이건 깃으로 버전관리를 하지 않고있다는 표시이다.  
앞서 만든 gitignore파일을 보면 venv 폴더가 버전관리를 하지않는 것으로 지정되어있다. 가상환경은 깃으로 버전관리를 하지 않은것이 좋다. window를 기준으로 개발을 하다 우분투나 리눅스 계열의 서버로 옮긴 다면 문제가 생길여지가 있기 때문이다.  
또한 gitignore에 한가지를 더 추가하자. 파이참에서 프로젝트에 설정한 내용을 담는 idea폴더인데 이것은 파이참에 의해 자동생성된 폴더이며 편의를 제공하기 위한 폴더이므로 버전 관리를 할 필요가 없다. 게다가 다른 컴퓨터에서 작업시 설정이 달라 문제가 생길 수 있다. gitignore파일을 열어 아랫줄에 .idea/ 를 추가해주자.

### 가상환경 실행하고 장고 설치하기
#### 가상환경 실행하기 
터미널에서 가상환경을 실행해보자. 터미널(cmder)에 에서 프로젝트 폴더로 들어가 venv\Scripts\activate을 입력하자.  
가상환경을 만든 후 pip install django를 설치한다.  
django startproject 프로젝트명 으로 프로젝트 폴더를 만들면 필요한 파일이 담긴 폴더가 생성된다.  
이제 python manage.py runserver로 서버를 실행시킨뒤  http://127.0.0.1:8000/  에 접속하면 성공이다.  

폴더안의 파일 중 urls.py / settings.py를 가장 많이 다룰 것이다. 전자는 사용자가 어떤 URL형식으로 접근했을 때 어떻게 웹을 작동시킬지 정리해 놓은 파일이다.  
후자는 설정을 담고있는 파일이다.

#### DB에 관리자 계정 생성하기
마이그레이션을 적용하기 전에 프로젝트는 제대로 작동하지 않는다. migration이란 DB에 적용시켜야하는 변화에 대한 기록이다.  
예를 들어 댓글 기능이 없던 블로그에 댓글 작성 기능을 추가 했다고 할 시, 댓글을 저장하기 위한 공간(테이블)이 필요하다. 이를 DB에 반영해야 서버를 실행했을 때 웹 사이트에 추가한 댓글 기능을 제대로 사용할 수 있다. 아직 기능이 없는 마이그레이션이 존재하는 이유는 장고에서 새 프로젝트를 생성할 때 DB에 기본적으로 필요한 테이블을 미련해주기 때문이다.

#### DB 생성하기
마이그레이션 적용을 위해 우선 단축키 Ctrl + C 를 눌러 서버를 중단시킨다.(원래는 python manage.py migrations 명령어를 먼저 지금은 장고가 만든 마이그레이션이 있으므로 생략) 이제 python manage.py migrate 명령어를 입력한다. 현재는 DB가 없어 db.sqlite3이라는 파일이 생성되고 그 안에 마이그레이션을 반영할 DB가 생성된다.
> SQLite3을 사용하는 이유
> SQLite3의 특징은 파일 하나로 관리하는 DB라는 점이다. 오라클,MY SQL,MongoDB 등 다른 DB는 복잡한 설치과정이 있고 백업을 위해 지식이 필요하다. 이에 비해 SQLite3는 파일 하나만 관리하면 된다는 장점이 있따. 물론 단점도 존재한다. 파일 기반의 DB이기 때문에 읽고 쓰기가 빈번한 대형 프로젝트에는 불리하다. 예를 들어 수천 수만명의 사용자가 몰리는 웹에선 적합하지 않으므로 고성능의 DB를 사용해야 한다.  

#### 관리자 계정 생성
이제 웹사이트의 관리자 계정(Super user)를 생성하자. 터미널에 python manage.py createsuperuser를 입력한다. 이메일과 비밀번호를 입력 해주자.  
이제 다시 python manage.py runserver로 서버를 실행 해주면 마이그레이션 오류가 없이 잘 실행되는 것을 볼 수 있다.  
이제 http://127.0.0.1:8000/admin/ 로 접속을 하면 관리자 계정과 비밀번호를 입력하는 창이 뜨고 입력을 하면 관리자 페이지에 접속할 수 있다.  
장고에서  이미 알아서 관리자 페이지를 생성한 것을 알 수 있다. 이제 여기에 기능을 추가해서 관리자 페이지를 확장할 수있다.

#### 깃에서 DB버전 관리하지 않도록 등록
혹시 db.sqlite3가 .gitignore에 없다면 등록해주자.  
> db.sqlite3를 깃으로 관리하지 않는 이유  
> 웹 개발을 하면 당분간 로컬 컴퓨에서 여러 테스트를 하면서 게시물이나 댓글을 작성해 볼것이다. 이러한 정보들이 db.sqlite3에 저장되고 모든 프로젝트 파일을 깃에 전송해야 하는데 테스트 과정의 정보까지 넘어가게 된다. 만약 이미 운영중인 서버에 db.sqlite3가 있다면 실제 방분자의 정보가 저장되어 있을 것이고 로컬 db.sqlite3와 충돌을 일으켜 오류가 발생하게된다.
